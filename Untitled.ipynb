```python
import os
import cv2
import torch
import torchvision.transforms as T
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
import glob

# Set your dataset folder path
image_folder = "./data/"
image_paths = sorted(glob.glob(os.path.join(image_folder, "*.jpg")))

# Define a torchvision transform
transform = T.Compose([
    T.Resize((480, 480)),  # Resize for easier visualization and speed
    T.ToTensor()
])

# Load and transform sample image
def load_image(index=0):
    image_path = image_paths[index]
    image = Image.open(image_path).convert("RGB")
    tensor_image = transform(image)
    
    # Show the image using matplotlib
    plt.figure(figsize=(6, 6))
    plt.imshow(np.array(image))
    plt.title(f"Loaded Image: {os.path.basename(image_path)}")
    plt.axis('off')
    plt.show()
    
    return image, tensor_image

# Test loading the first image
image, tensor_image = load_image(0)
```

```python
from transformers import DetrImageProcessor, DetrForObjectDetection
import torch

# Load the pretrained DETR processor and model
processor = DetrImageProcessor.from_pretrained("facebook/detr-resnet-50")
model = DetrForObjectDetection.from_pretrained("facebook/detr-resnet-50")

# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Prepare inputs and predict
inputs = processor(images=image, return_tensors="pt").to(device)
outputs = model(**inputs)
```
